# 👤 Person 3 — Algorithm + Formula

## ⚙️ Key Equations (Core of PSO)

### Velocity Update

v = wv + c_1 r_1 (pbest - x) + c_2 r_2 (gbest - x)

### Position Update

x = x + v

### 🧠 Meaning of Terms (simple explanation)

* **v** → velocity (how fast particle moves)
* **x** → current position (solution)
* **w** → inertia (keeps moving in same direction)
* **c₁, c₂** → learning factors (how much it learns)
* **r₁, r₂** → random numbers (add randomness)
* **pbest** → particle’s best solution so far
* **gbest** → best solution found by the swarm

👉 Idea:
Each particle moves based on:

* its own experience (**pbest**)
* group knowledge (**gbest**)

---

## 🔄 Step-by-Step Algorithm

### 1. Initialize Particles

* Create many particles (solutions)
* Assign:

  * random position (**x**)
  * random velocity (**v**)

---

### 2. Evaluate Fitness

* Use a **fitness function**
* Example:

  * accuracy (ML)
  * error (minimization problem)

👉 This tells how good each particle is

---

### 3. Update pbest & gbest

* For each particle:

  * If current solution is better → update **pbest**
* Among all particles:

  * Best one becomes **gbest**

---

### 4. Update Velocity & Position

* Use formulas:

  * update **velocity**
  * update **position**

👉 Particles “move” toward better solutions

---

### 5. Repeat

* Repeat steps:

  * Evaluate → Update → Move
* Stop when:

  * max iterations reached
  * or solution is good enough

---

## 🎯 Simple Way to Say in Presentation

“Each particle tries a solution, remembers its best result, and learns from the best in the group. Step by step, all particles move toward the optimal solution.”
